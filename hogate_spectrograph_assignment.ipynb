# Spectral analysis of sounds


```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
#assign sound files to variables
sound1 = r'C:\Users\Devin\Documents\23_24Spring\GES371\data\nri-Thunder.wav'
#source1: https://www.nps.gov/subjects/sound/sounds-thunder.htm
sound2 = r'C:\Users\Devin\Documents\23_24Spring\GES371\data\lynyrd-skynyrd-free-bird-(guitar-solo)-made-with-Voicemod.wav'
```

```python
#convert the sound files to an array
sound1y, sound1sr = librosa.load(sound1)
thunder = librosa.feature.melspectrogram(y=sound1y, sr=sound1sr)

sound2y, sound2sr = librosa.load(sound2)
freebird = librosa.feature.melspectrogram(y=sound2y, sr=sound2sr)
```

```python
thunder
```

```python
#time series? of 1st sound file
#thundersin = 
plt.plot(thunder)
plt.xlim(0,46)
```

```python
#make the spectrogram of the first sound(thunder)
fig, ax = plt.subplots()
thunder_dB = librosa.power_to_db(thunder, ref=np.max)
img = librosa.display.specshow(thunder_dB, x_axis='time',
                         y_axis='mel', sr=sound1sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Thunder spectrogram')
plt.xlim(0,20)
```

```python
#time series? of 2nd sound file
plt.plot(freebird)
plt.xlim(0,75)
```

```python
#spectrogram of 2nd sound file
fig, ax = plt.subplots()
freebird_dB = librosa.power_to_db(freebird, ref=np.max)
img = librosa.display.specshow(freebird_dB, x_axis='time',
                         y_axis='mel', sr=sound2sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Freebird Solo spectrogram')
plt.xlim(0,8)
```
