# Historical Energy Use in Hawaii 

```python
#import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#get data source
#source: https://energy.hawaii.gov/what-we-do/clean-energy-vision/transportation/#/analyze?fuel=ELEC&country=US&region=US-HI&show_map=true
data_src = open(r"C:\Users\Devin\Documents\23_24Spring\GES371\data-EhCSP.csv")
HI_energy = pd.read_csv(data_src, header=0)
HI_energy.head()
```

```python
#format the data
HI_energy.rename(columns={'ï»¿Year':'Year'}, inplace=True)
```

```python
HI_energy.columns
```

```python
#name title
plt.title('Energy use in Hawaii by Sector')

#plot the thing for each sector
plot1 = plt.plot(HI_energy['Year'], HI_energy['Residential'], color='red', label='Residential')
plot2 = plt.plot(HI_energy['Year'], HI_energy['Commercial'], color='green', label='Commercial')
plot3 = plt.plot(HI_energy['Year'], HI_energy['Industrial'], color='blue', label='Industrial')
plot4 = plt.plot(HI_energy['Year'], HI_energy['Transportation'], color='purple', label='Transportation')

#graph style
plt.xlabel('Year')
plt.ylabel('Energy use (Trillions BTU)')

allplot = plot1 + plot2 + plot3 + plot4
labels = [i.get_label() for i in allplot]
plt.legend(labels, loc=0, fontsize='small')
```

# The chart shows the changes in Energy use by sector in Hawaii
This chart shows the trends of energy use in Hawaii, separated by sectors, measured in trillions of BTU. In Hawaiʻi, the per capita use of residentail energy in 2021 was 26 MMBtu per person (US Energy Information Administration). The transportation sector is by far the biggest sector by use. This is becuase it includes the use of jet fuel by airliners. The industrial sector shows an overall decrease in its usage of energy, while Residentail and Commercial sectors show a slight positive slope. This could be due to the increase of population over time.
