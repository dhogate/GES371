# Analysis of HOT CO2 data
data come from the HOT site, http://hahana.soest.hawaii.edu/hot, in the CO2 link

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data
# the data are online at
url = 'https://hahana.soest.hawaii.edu/hot/hotco2/HOT_surface_CO2.txt'

# read data from that URL, note that the columns are not comma separated, but white space
# use delim_whitespace=True to demark columns
# note there is some header info at the top, so need to skip, in this case 7 lines
# header=7
HOT = pd.read_csv(url, delim_whitespace=True, header=7)
```

```python
HOT
```

```python
# Note either by looking at the table or making a plot:
#HOT['pHmeas_insitu'].plot(kind='line')
# missing values seem to be set to -999.0
# we need to remove, and replace with NaN's
HOT_nomissing = HOT.replace(-999.0, np.nan)
```

```python
HOT_nomissing['temp'].plot(kind='line')
```

```python
# make more reasonable date for the x-axis
# python has many solutions for dates; here we'll use the pandas "to_datetime"
# note here we are over-wrting the column with new values
HOT_nomissing['date'] = pd.to_datetime(HOT_nomissing['date'])
```

```python
plt.plot(HOT_nomissing['date'],HOT_nomissing['temp'])
```

make a three-panel plot, temperature, salinity and total akalinity

```python
plt.subplots(nrows=3,ncols=1,figsize=(14,10))

plt.subplot(3,1,1)
plt.plot(HOT_nomissing['date'],HOT_nomissing['temp'])
plt.title('HOT surface temperature')
plt.grid()

plt.subplot(3,1,2)
plt.plot(HOT_nomissing['date'],HOT_nomissing['sal'])
plt.title('HOT surface salinity')
plt.grid()

plt.subplot(3,1,3)
plt.plot(HOT_nomissing['date'],HOT_nomissing['TA'])
plt.title('HOT surface alkalinity')
plt.grid()
```

```python
plt.scatter(HOT_nomissing['temp'],HOT_nomissing['DIC'])
```
