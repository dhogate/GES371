```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cf
```

```python
#make map and size
plt.figure(figsize=(18,12))

#set map type and scope
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=0))
ax.set_extent([-180,20,-75,75], crs=ccrs.PlateCarree())

#add map features and image
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.OCEAN)
ax.add_feature(cf.BORDERS)
ax.stock_img()

#add map lines and lat&long on axes
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xlabels_style = {'size':20}
gl.ylabels_style = {'size':20}

#Locations and their coords
LOC_HONO = (-157.7, 19.5)
LOC_SJC = (-122.4, 36.2)
LOC_DUB = (-6.3-.5, 52.7)
LOC_LONDON = (-3, 51.5)
LOC_STALOHA = (-158, 22.5)

#add locations to map w/ 'annotate', string, place, color, etc.
ax.annotate('* Honolulu', xy=LOC_HONO, 
           xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
           color='black', size='20', ha='left', va='center')
ax.annotate('* San Jose', xy=LOC_SJC, 
           xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
           color='black', size='20', ha='left', va='center')
ax.annotate('Dublin*', xy=LOC_DUB, 
           xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
           color='black', size='20', ha='right', va='center')
ax.annotate('* London', xy=LOC_LONDON, 
           xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
           color='black', size='20', ha='left', va='center')
ax.annotate('* Station Aloha', xy=LOC_STALOHA, 
           xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
           color='blue', size='15', ha='left', va='center')
```
